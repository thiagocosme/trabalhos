### Questão 1

Denomina-se épsilon da máquina o menor número que somado a 1 produza resultado diferente de 1, ou seja, o menor valor que não é arredondado para baixo.

```python
eps = 1.0
n = 0

while (eps+1) > 1.0:
    
    eps = eps/2
    n = n +1
    
print ("\n \t Épsilon da máquina: ", 2*eps)
```

O valor obtido num computador de 64 bit foi de 2.22044604925e-16. O mesmo resultado foi obtido rodando o programa em um smartphone Moto G5 S e em um Moto G3.


Vale lembrar que, no último laço que o algoritmo entra, o valor do épsilon é dividido por dois e então a soma 'eps + 1' retorna o valor 1, quebrando o laço. Assim o valor real do épsilon será o dobro do ultimo valor atribuído à variável 'eps'.


### Questão 2:

Considerando a expressão,

$$\cfrac{e^{1/x}}{1+e^{1/x}} ~~~~(Eq. 1)$$

Dividindo tanto a expressão do denominador quanto a do numerador por $e^{1/x}$ temos:

$$\cfrac{e^{1/x}}{1+e^{1/x}} \div \cfrac{e^{1/x}}{e^{1/x}} = \cfrac {\cfrac{e^{1/x}}{e^{1/x}}}{\cfrac{1+e^{1/x}}{e^{1/x}}} = \cfrac{1}{\cfrac{1}{e^{1/x}}+1} = \boxed {~~~~\cfrac{1}{e^{-1/x}+1}~~~~} ~~~~(Eq. 2)$$



Agora usando o python para calcular a expressão 1 para valores pequenos de x temos:

```python
import math                        # biblioteca de matemática para usar a função exp()

print( "\n")
print ("Cálculo via Eq. 1\n")

for i in range (300, 0, -1):       # range (inicio, fim, passo)
    
    x = 0.0001*i
    print (i,"\t x = %.20f" % x, "| \t f(x) =", math.exp(1.0/x)/(1 + math.exp(1.0/x)))   
    
```

Há um erro de overflow para a função exponencial quando x = 0.0014.

### Questão 3

O número $e$ pode ser defnido pela série 
$$ e= \sum\limits_{n=0}^{\infty} \cfrac{1}{n!} = \cfrac{1}{0!} + \cfrac{1}{1!} + \cfrac{1}{2!} + \cfrac{1}{3!} +~~ ... ~~+ \cfrac{1}{n!} \\ e = 1 + 1 + \cfrac{1}{2} + \cfrac{1}{6} +\cfrac{1}{24} +~~ ... ~~+ \cfrac{1}{n!}$$

Assim, podemos escrever uma estrutura de repetição em python, usando duas variáveis iniciais 'e' e 'e_prox', tais que o primeiro valor de 'e' será zero e o 'e_prox' será o primeiro termo da série. Definimos a função que retorna o valor de $\cfrac{1}{n!}$ para cada valor de 'n' e adicionamos às variáveis, fazendo 'n' para 'e' e 'n+1' para 'e_prox', assim temos o valor de $e$ para cada termo 'n + 1' onde truncamos a série e o anterior 'e' para calcularmos o erro relativo.

Abaixo segue o código em python.

```python
import math

e = 0                                   # damos o valor do primeiro termo (1/0!) a 'e',
e_prox = 1                              # zero para o termo anterior
n = 0                                   # e iniciamos a série a partir de n = 0
erro= 10

f = lambda n: 1.0/math.factorial(n)


while erro > 0.0001:

    e = e + f(n)
    e_prox = e_prox + f(n+1)            # note que o último valor da nossa série é e_prox, logo é o nosso valor de e que procuramos
    n = n + 1
    
    erro = abs((e_prox - e)/e_prox)
    
print ("\n O valor de 'e' com erro relativo menor que 0.0001 é: ", e_prox)
```

### Questão 4

A fórmula de Leibniz aproxima o número pi pelo somatório: 

$$\cfrac{\pi}{4} = \sum \limits_{n=0}^{\infty} \cfrac{(-1)^n}{2n + 1}$$

Portanto, podemos fazer um programa em python com 50 repetições, onde o programa retornará o valor mais próximo de pi, nesse numero finito de laços.

Segue abaixo o programa em python:


```python
pi = 0                                   # damos o valor do primeiro termo (1/0!) a 'e',
pi_prox = 0                              # zero para o termo anterior
                                   

for n in range(50):                      # e iniciamos a série a partir de n = 1 até n = 50

    pi = pi + (-1.0)**n/(2*n+1)
    n = n + 1
    
    #print n,pi
    
pi = 4*pi                                # O loop nos fornece pi/4, logo o pi verdadeiro será 4*pi
print ("\n Pi é aprox. igual a",pi, "usando 50 termos da série.")
```

O número que o programa retornou é apenas aproximado do 'pi real' pois não houveram laços suficientes para completar o valor exato.


### Questão 5

Foi observado que para função $ln(1+x)$ a partir da segunda derivada a função segue o padrão: $(-1)^{n+1}(n-1)!\cdot(x+1)^{-n}$. Foi criada a função f da série de Taylor em função de n correspondente ao enésimo termo:

$$ f^{(n)}(x_0) \cfrac{(x - x_0)^n}{n!} = \left [ ~~ (1 + x_0)^{-n}(-1)^{n+1}(n-1)! ~~ \right ] ~~ \cfrac{(x-x_0 )^n }{n!} $$   
$$  = \boxed {~~~~(-1)^{n+1}(1 + x_0)^{-n} ~~ \cfrac{(x-x_0 )^n}{n}~~~~}$$

onde a cada laço é somado com o termo anterior da série. 
Segue abaixo o programa em python:

```python
import math

x = -0.2
x0 = 0
erro = 0.1
n = 1

p = math.log(x0+1)                        # Iniciamos a série com o valor do primeiro termo = f(x0 + 1) 
p_prox = 0
                                          # f será a repetição que foi observada a partir do segundo termo

f = lambda n: (-1)**(n+1) * (x0+1)**(-n) * ((x-x0)**n)/n

while erro > 0.0001:                      # Através do erro que o loop do while permanece até a tolerância estabelecida
    
    p = p + f(n)
    p_prox = p + f(n+1)
    erro = abs(p_prox - p)
    n = n + 1
    print("Erro: ",erro,"\t Valor de P:",p,"\t Valor do Prox Termo:",p_prox)
    
n = n + 1

print ("\n\nPara que o erro absoluto seja menor que 0.0001, são necessários",n,"termos no polinômio de Taylor")
    
```

A variável p é o primeiro termo da série e a variável p_prox é o termo posterior. O loop while funciona a partir do erro absoluto entre o valor da variável |p_prox - p|, do qual a tolerância deve ser 0.0001.
O grau do polinômio de Taylor no programa começa n = 1 pois ja temos declarado o valor do primeiro termo (n = 0) na variável "p" e a cada loop o grau do polinômio é somado mais 1. Assim foi adicionado +1 ao 'n'.
